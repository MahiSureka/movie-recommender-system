```python
import pandas as pd

# source: https://www.kaggle.com/datasets/rishidamarla/bollywood-movies-dataset
# Bollywood Movies Dataset
# 1698 Hindi Movies from 2005-2017

bollywood = pd.read_csv("../data/Data for repository.csv")
```

```python
# keep only useful columns

bollywood = bollywood[
    [
        "Movie Name",
        "Genre",
        "Lead Star",
        "Director",
        "Music Director",
        "Whether Franchise",
        "Whether Remake",
        "Release Period"
    ]
]
```

```python
# check missing values

bollywood.isnull().sum()
```

```python
bollywood.dropna(inplace=True)
```

```python
# rename the movie column

bollywood.rename(
    columns={"Movie Name": "title"},
    inplace=True
)
```

```python
# convert to strings

for col in bollywood.columns:
    bollywood[col] = bollywood[col].astype(str)
```

```python
# make tags

bollywood["tags"] = (
    bollywood["Genre"]
    + " "
    + bollywood["Lead Star"]
    + " "
    + bollywood["Director"]
    + " "
    + bollywood["Music Director"]
    + " "
    + bollywood["Whether Franchise"]
    + " "
    + bollywood["Whether Remake"]
    + " "
    + bollywood["Release Period"]
)
```

```python
# Keep only title and tags

new_df = bollywood[["title", "tags"]]
```

```python
# lowercase

new_df["tags"] = new_df["tags"].str.lower()
```

```python
# vectorize 

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(
    max_features=5000,
    stop_words="english"
)

vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
# similarity

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
# recommendation functon 

def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]

    distances = sorted(
        list(enumerate(similarity[movie_index])),
        key=lambda x: x[1],
        reverse=True
    )

    for i in distances[1:6]:
        print(new_df.iloc[i[0]]["title"])
```

```python
recommend("Yeh Jawaani Hai Deewani")
```

```python
recommend("Housefull")
```

```python
recommend("Taare Zameen Par")
```

```python
recommend("3 Idiots")
```
